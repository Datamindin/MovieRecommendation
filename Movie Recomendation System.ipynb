```python
import spacy

nlp = spacy.load("en_core_web_sm")

def preprocess(text):
    filtered_token = []
    doc = nlp(text)
    
    for token in doc:
        if token.is_stop or token.is_punct:
            continue
        filtered_token.append(token.lemma_)
        
    return " ".join(filtered_token)
        
```

```python
preprocess("While Frodo and Sam edge closer to Mordor with the help of the shifty Gollum, the divided fellowship makes a stand against Sauron's new ally, Saruman, and his hordes of Isengard.")
```

```python
df['Processed_Overview'] = df['Overview'].apply(preprocess)
```

```python
df.rename(columns = {'Movie Title':'title'}, inplace = True)
```

```python
def create_soup(data):
    # Creating a simple text for countvectorizer to work with
    att = data['title'].lower()
    for i in data[1:]:
        att = att + ' ' + str(i)
    return att
 
model_data=df.copy()
model_data=model_data[['title','Certificate','Genre','Director','Star1','Star2','Star3','Star4','Processed_Overview']]
model_data['soup']=model_data.apply(create_soup,axis=1)
```

```python
model_data['soup'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
count = CountVectorizer()
count_matrix = count.fit_transform(model_data['soup'])
cosine_sim2 = cosine_similarity(count_matrix)
```

```python
cosine_sim2
```

```python
def recommend(movie):
    index = model_data[model_data['title'] == movie].index[0]
    distances = sorted(list(enumerate(cosine_sim2[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(model_data.iloc[i[0]].title)
        
```
